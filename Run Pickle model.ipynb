```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import pickle
```

```python
url = "https://biostat.app.vumc.org/wiki/pub/Main/DataSets/titanic3.xls"
df = pd.read_excel(url)
df.head()
```

```python
df = df.drop(columns=["name","ticket","home.dest","boat", "body", "cabin",])
df.head()
```

```python
X = df.drop(columns="survived")
y = df.survived
```

```python
X
```

```python
filename = 'bestModel.sav'
model= pickle.load(open(filename,'rb')) #'rb' means to read bites
model
```

```python
model.score(X,y)
```

```python
model.predict(X.head(1))
```

```python
model.predict_proba(X.head(1))
```
